## Test 6

```python
def fix_columns_name(df):
    return [column[0] if column[0] in ['caso', 'tenure', 'max_iter', 'sol_ini'] else column[0] + "_" + column[1] for column in df.columns]
```

```python
raw_data = pd.read_csv("teste6.csv")
```

```python
test_data = raw_data.drop(list(range(217, 241)), axis=0).reset_index()
```

```python
test_data[test_data.tempo > 310]
```

```python
test_data2 = test_data[test_data.tempo < 310].drop(columns=['index', 'num_teste'])
```

```python
test_data3 = test_data2.groupby(['caso', 'tenure', 'max_iter', 'sol_ini']).agg({'custo': ['mean', 'min', np.std], 'tempo': ['mean'], 'gap': ['mean', 'min']}).reset_index()
```

```python
test_data4 = test_data2.groupby(['caso', 'tenure', 'max_iter', 'sol_ini']).agg({'custo': ['mean', 'min'], 'gap': ['mean']}).reset_index()
```

```python
top = test_data3.sort_values('gap_mean').drop_duplicates(subset=['caso', 'tenure', 'max_iter'], keep='first')
```
